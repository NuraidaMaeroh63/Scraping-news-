```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm.auto import tqdm
import json
```

```python
url = 'https://thestandard.co/page/'
```

```python
bs = []
page = 1

while page <= 10:
    data = requests.get(url + str(page) + '/?s=7-eleven&search')
    soup = BeautifulSoup(data.content, 'html.parser')
    bs.append(soup)
    page += 1
data
```

```python
title_list = []
urls = []
author_list = []
desc_list = []
```

```python
for i in range(len(bs)):
    for t in bs[i].find_all('div', {'class' : 'col-md-8'}):
        urls.append(t.find('a', href=True)['href'])
        title_list.append(t.find('h3',{'class':'news-title'}).find('a').text.replace('\n',''))
```

```python
for link in urls:
    page_url = requests.get(link)
    b = BeautifulSoup(page_url.content, 'html.parser')
    for a in b.findAll('div', {'class' : 'meta-author'}):
        author_list.append(a.find('a').text)
        
    for d in b.findAll('div', {'class' : 'entry-content'}):
        des = d.text.replace('\n','').replace('\xa0','')
        desc_list.append(des)
```

```python
table = pd.DataFrame([title_list, author_list, desc_list, urls]).transpose()
table.columns = ['title','author','description','source_url']
table.head(10)
```

```python
table.dtypes
```

```python
table.to_csv('scrap_7eleven_TheStandard_news.csv', index = False)
```

```python
# Convert CSV to JSON

import csv 
import json 

def csv_to_json(csvFilePath, jsonFilePath):
    jsonArray = []
      
    #read csv file
    with open(csvFilePath, encoding='utf-8') as csvf: 
        #load csv file data using csv library's dictionary reader
        csvReader = csv.DictReader(csvf) 

        #convert each csv row into python dict
        for row in csvReader: 
            #add this python dict to json array
            jsonArray.append(row)
  
    #convert python jsonArray to JSON String and write to file
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf: 
        jsonString = json.dumps(jsonArray, indent=4)
        jsonf.write(jsonString)
          
csvFilePath = r'scrap_7eleven_TheStandard_news.csv'
jsonFilePath = r'scrap_7eleven_TheStandard_news.json'
csv_to_json(csvFilePath, jsonFilePath)
```

# NER (NameEntitiesRegcognition)

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("Davlan/xlm-roberta-base-ner-hrl")
model = AutoModelForTokenClassification.from_pretrained("Davlan/xlm-roberta-base-ner-hrl")
nlp = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)
```

```python
import re
def post_process(preds):
    data = []
    for i in range(len(preds)): 
        a = preds[i]
        text = a.get('word')
        for j in range(i, len(preds)):
            b = preds[j]
            if b.get('start') == a.get('end'):
                text += b.get('word')
                a = b
        data.append({
            'type': a.get('entity'),
            'text': text
        })
    tmp = []
    for elm in data:
        text = elm.get('text')
        if bool(re.match("[ก-๛]|▁", text)):
            elm.update({'text': text.replace('▁', ' ')})
            tmp.append(elm)
    return tmp
```

```python
text = data[0]['description']
text
```

```python
output = nlp(text)
output
```

```python
post_process(output)
```

# Process news to spo

```python
spo = []
for news in tqdm(data):
    description = news.get('description') 
    if description == None:
        continue
    ner = nlp(description)
    processed = post_process(ner)
    for kw in processed:
        spo.append({
            'subject': news.get('title', ""),
            'predicate': kw.get('type', ""),
            'object': kw.get('text', "")
        })
```

```python
df = pd.DataFrame(spo)
df['predicate'] = df['predicate'].apply(lambda x: x[2:])
df
```
